# Previsão do Preço de Casas com Regressão Linear Múltipla

Dataset: [House Sales in King County, USA](https://www.kaggle.com/harlfoxem/housesalesprediction)

```python
import pandas as pd
base = pd.read_csv('data/house-prices.csv')
base.head()
```

```python
base.columns
```

```python
colunas_usadas = ['price', 'bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long']
```

```python
colunas_usadas
```

```python
base = pd.read_csv('data/house-prices.csv', usecols=colunas_usadas)
```

```python
base.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler_x = MinMaxScaler()
```

```python
base[['bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long']] = scaler_x.fit_transform(base[['bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long']])
```

```python
base.head()
```

```python
scaler_y = MinMaxScaler()
```

```python
base[['price']] = scaler_y.fit_transform(base[['price']])
```

```python
base.head()
```

```python
X = base.drop('price', axis=1)
```

```python
y = base.price
```

```python
X.head()
```

```python
y.head()
```

```python
type(X)
```

```python
type(y)
```

```python
previsores_colunas = colunas_usadas[1:]
previsores_colunas
```

```python
import tensorflow as tf
```

```python
colunas = [tf.feature_column.numeric_column(key=c) for c in previsores_colunas]
```

```python
colunas
```

```python
from sklearn.model_selection import train_test_split
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(X, y, test_size=0.3)
```

```python
X_treinamento.shape
```

```python
y_treinamento.shape
```

```python
funcao_treinamento = tf.estimator.inputs.pandas_input_fn(x=X_treinamento, y=y_treinamento, batch_size=32, num_epochs=None, shuffle=True)
```

```python
funcao_teste = tf.estimator.inputs.pandas_input_fn(x=X_teste, y=y_teste, batch_size=32, num_epochs=10000, shuffle=False)
```

```python
regressor = tf.estimator.LinearRegressor(feature_columns=colunas)
```

```python
regressor.train(input_fn=funcao_treinamento, steps=10000)
```

```python
metricas_treinamento = regressor.evaluate(input_fn=funcao_treinamento, steps=10000)
```

```python
metricas_teste = regressor.evaluate(input_fn=funcao_teste, steps=10000)
```

```python
metricas_treinamento
```

```python
metricas_teste
```

```python
funcao_previsao = tf.estimator.inputs.pandas_input_fn(x=X_teste, shuffle=False)
```

```python
previsoes = regressor.predict(input_fn=funcao_previsao)
```

```python
list(previsoes)
```

```python
valores_previsoes = [p['predictions'] for p in regressor.predict(input_fn=funcao_previsao)]
```

```python
valores_previsoes
```

```python
import numpy as np
valores_previsoes = np.asarray(valores_previsoes).reshape(-1, 1)
```

```python
valores_previsoes
```

```python
valores_previsoes = scaler_y.inverse_transform(valores_previsoes)
```

```python
valores_previsoes
```

```python
y_teste
```

```python
y_teste.shape
```

```python
y_teste_as_matrix = y_teste.values.reshape(-1, 1)
```

```python
y_teste_as_matrix.shape
```

```python
y_teste_as_matrix = scaler_y.inverse_transform(y_teste_as_matrix)
```

```python
y_teste_as_matrix
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mae = mean_absolute_error(y_teste_as_matrix, valores_previsoes)
```
